```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm
```

```python
sample_space = ['H', 'T']
prob_H = 1 / len(sample_space)
prob_T = 1 / len(sample_space)

print("Probability of Head:", prob_H)
print("Probability of Tail:", prob_T)
```

```python
n = 10   # number of trials
p = 0.5  # probability of success

x_binom = np.arange(0, n+1)
y_binom = binom.pmf(x_binom, n, p)
```

```python
plt.figure()
plt.bar(x_binom, y_binom)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.grid()
plt.show()
```

```python
lam = 3  # average rate

x_pois = np.arange(0, 10)
y_pois = poisson.pmf(x_pois, lam)

plt.figure()
plt.bar(x_pois, y_pois)
plt.title("Poisson Distribution (λ=3)")
plt.xlabel("Events")
plt.ylabel("Probability")
plt.grid()
plt.show()
```

```python
mu = 0     # mean
sigma = 1  # standard deviation

x_norm = np.linspace(-4, 4, 100)
y_norm = norm.pdf(x_norm, mu, sigma)
```

```python
plt.figure()
plt.plot(x_norm, y_norm)
plt.title("Normal Distribution (μ=0, σ=1)")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid()
plt.show()
```
